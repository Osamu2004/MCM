## 最小-最大标准化（Min-Max Scaling）：

```python
def Min_Max_sum(data_pd):    
    data = np.asarray(data_pd)
    _, m = data.shape
    # 求均值
    mean_value = np.mean(data,axis = 0)
    # 求标准差
    std_value = np.std(data,axis = 0)
    data_s = (data - mean_value) / std_value
    t = np.sum(data_s,axis = 1)
    return t

sample1_r = Min_Max_sum(data_pd1)
sample1_w =  Min_Max_sum(data_pd2)
sample1 = np.concatenate((sample1_r, sample1_w))
sample2_r = Min_Max_sum(data_pd3)
sample2_w = Min_Max_sum(data_pd4)
sample2 = np.concatenate((sample2_r, sample2_w))
sample1
```

```python
def z_score(data_pd):
    data = np.asarray(data_pd)
    _, m = data.shape
    data_max = data.max(axis = 0).reshape(1,m) #定义最大值
    data_min = data.min(axis = 0).reshape(1,m) #定义最小值
    data_s = (data - data_min) / (data_max - data_min)
    t = np.sum(data_s,axis = 1)
    return t
sample1_r = z_score(data_pd1)
sample1_w =  z_score(data_pd2)
sample1 = np.concatenate((sample1_r, sample1_w))
sample2_r = z_score(data_pd3)
sample2_w = z_score(data_pd4)
sample2 = np.concatenate((sample2_r, sample2_w))
sample1
```

```python
from scipy import stats
def Shapiro_Wilk(data,alpha):
    statistic, p_value = stats.shapiro(data)
    # 输出检验结果
    print("Shapiro-Wilk statistic:", statistic)
    print("p-value:", p_value)
    # 判断是否符合正态分布
    if p_value > alpha:
        print("样本数据符合正态分布")
    else:
        print("样本数据不符合正态分布")
Shapiro_Wilk(sample1,0.10)
Shapiro_Wilk(sample2,0.10)
```

```python
from scipy import stats

# 假设你有两个独立样本数据 sample1 和 sample2，可以是两个列表或NumPy数组
# 执行独立样本t检验
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# 输出检验结果
print("T-statistic:", t_statistic)
print("p-value:", p_value)

# 判断是否存在显著差异
alpha = 0.05  # 设置显著性水平
if p_value < alpha:
    print("样本之间存在显著差异")
else:
    print("样本之间不存在显著差异")
```

## z-score标准化（Standardization）：

```python
def Min_Max_sum(data_pd):    
    data = np.asarray(data_pd)
    _, m = data.shape
    # 求均值
    mean_value = np.mean(data,axis = 0)
    # 求标准差
    std_value = np.std(data,axis = 0)
    data_s = (data - mean_value) / std_value
    t = np.sum(data_s,axis = 1)
    return t

def z_score(data_pd):
    data = np.asarray(data_pd)
    _, m = data.shape
    data_max = data.max(axis = 0).reshape(1,m) #定义最大值
    data_min = data.min(axis = 0).reshape(1,m) #定义最小值
    data_s = (data - data_min) / (data_max - data_min)
    t = np.sum(data_s,axis = 1)
    return t
```
