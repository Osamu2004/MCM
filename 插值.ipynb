```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicHermiteSpline

plt.rcParams['font.family'] = 'SimSun'  # 设置全局字体为SimSun

# 原始数据点，使用sin(x)作为函数值
x = np.array([1, 2, 3, 4, 5])
y = np.sin(x)  # 使用sin(x)作为函数值
dy_dx = np.cos(x)  # sin(x)的导数值为cos(x)

# 创建三次埃尔米特插值
spline = CubicHermiteSpline(x, y, dy_dx)

# 在插值点上进行插值
x_interp = np.linspace(1, 5, 100)
y_interp = spline(x_interp)

# 绘制原始数据和插值结果
plt.scatter(x, y, label='原始数据', color='blue')
plt.plot(x_interp, y_interp, label='三次埃尔米特插值', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('三次埃尔米特插值')
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

plt.rcParams['font.family'] = 'SimSun'  # 设置全局字体为SimSun

# 原始数据点，使用sin(x)作为函数值
x = np.array([1, 2, 3, 4, 5])
y = np.sin(x)  # 使用sin(x)作为函数值

# 创建三次样条插值
spline = CubicSpline(x, y)

# 在插值点上进行插值
x_interp = np.linspace(1, 5, 100)
y_interp = spline(x_interp)

# 绘制原始数据和插值结果
plt.scatter(x, y, label='原始数据', color='blue')
plt.plot(x_interp, y_interp, label='三次样条插值', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('三次样条插值')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 定义原始数据点
x = np.array([1, 2, 3, 4, 5])
y = np.array([0.5, 0.8, 2.3, 3.5, 1.2])  # 函数值

# 创建插值函数对象
# kind参数指定插值方法，可以选择'linear'（线性插值）、'nearest'（最近邻插值）、'quadratic'（二次插值）或'cubic'（三次插值）
interp_func = interp1d(x, y, kind='cubic')

# 在插值点上进行插值
x_interp = np.linspace(1, 5, 100)  # 在原始数据点的范围内生成100个插值点
y_interp = interp_func(x_interp)  # 返回对应的插值y值

# 绘制原始数据和插值结果
plt.scatter(x, y, label='原始数据', color='blue')
plt.plot(x_interp, y_interp, label='三次插值', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('三次插值')
plt.grid(True)
plt.show()
```
