```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
plt.rcParams['font.family'] = 'SimSun'  # 设置全局字体为SimSun
# 示例数据
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 3.8, 6.2, 8.1, 9.7, 12.5])

# 将x变为二维数组
X = x[:, np.newaxis]

# 创建线性回归模型
model = LinearRegression()

# 拟合数据
model.fit(X, y)

# 得到斜率和截距
slope = model.coef_[0]
intercept = model.intercept_

# 打印拟合得到的直线的方程
print(f"拟合得到的直线方程为：y = {slope:.2f}x + {intercept:.2f}")

# 可视化拟合结果
plt.scatter(x, y, color='blue', label='数据点')
plt.plot(x, model.predict(X), color='red', label='拟合线')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('线性回归拟合')
plt.show()
```
