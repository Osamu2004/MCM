## 线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.family'] = 'SimSun'  # 设置全局字体为SimSun
xlsx = pd.ExcelFile(r"C:\Users\86158\Desktop\a积累代码\清风数学建模课件和代码（全套下载后请解压)\正课配套的课件和代码\第4讲.拟合\代码和例题数据\data1.xlsx")
data_pd = pd.read_excel(xlsx, 'Sheet1')
x = np.array(data_pd.iloc[:, 0])
y = np.array(data_pd.iloc[:, 1])
```

```python
plt.scatter(x, y, marker='o', linestyle='-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y against x')
plt.grid(True)
plt.show()
```

```python
n = len(x)
k = (n*np.sum(x*y) - np.sum(x)*np.sum(y))/(n*np.sum(x*x) - np.sum(x)*np.sum(x))
b = (np.sum(x*x)*np.sum(y) - np.sum(x)*np.sum(x*y))/(n*np.sum(x*x)-np.sum(x)*np.sum(x))

x_fit = np.arange(round(x.min()) - 1, round(x.max()) + 1, 0.001)
y_fit = x_fit * k + b
print(k)
print(b)

plt.scatter(x, y, label='原始数据', color='blue')
plt.plot(x_fit, y_fit, label='线性拟合', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('线性拟合')
plt.grid(True)
plt.show()
```
