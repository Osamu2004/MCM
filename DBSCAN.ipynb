```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np

# Generate moon-shaped data
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# Scale the data to have zero mean and unit variance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create DBSCAN object
dbscan = DBSCAN(eps=0.6)

# Fit DBSCAN to the scaled data and predict clusters
clusters = dbscan.fit_predict(X_scaled)

# Define colormap for the plot
cm2 = ListedColormap(['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2'])

# Plot the data points with assigned clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, s=60, cmap=cm2)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
